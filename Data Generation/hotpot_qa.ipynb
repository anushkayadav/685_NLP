```python
import numpy as np
# Function to extract the supporting sentences based on the supporting facts
def extract_supporting_sentences(row):
    contexts = row['context']
    facts = row['supporting_facts']
    result = []
    for title, sent_ids in zip(facts['title'], facts['sent_id']):
        # Ensure sent_ids is iterable
        sent_ids = np.atleast_1d(sent_ids)
        # Find index of the title in the context titles
        idx = np.where(contexts['title'] == title)[0][0]
        # Retrieve the sentences using the sent_ids, ensuring that no invalid indices are used
        valid_sent_ids = [sid for sid in sent_ids if sid < len(contexts['sentences'][idx])]
        sentences = contexts['sentences'][idx][valid_sent_ids]
        sentences_text = " ".join(sentences)  # Combine multiple sentences if there are more than one
        result.append(f"{title} > {sentences_text}")
    return ' | '.join(result)

# Apply the function to each row in the DataFrame
train_df['W'] = train_df.apply(extract_supporting_sentences, axis=1)

# Print the resulting DataFrame
print(train_df['W'])
```
